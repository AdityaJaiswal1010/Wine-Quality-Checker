```python
def find_outliner_limits(column):
    Q1,Q3=stock_df[column].quantile([.25,.75])
    IQR=Q3-Q1
    
    high=Q3+(1.5*IQR)
    return (high,IQR)


high_vol,IQR=find_outliner_limits('Volume')
print("Volume : \n","Upper Limit : ",high_vol,"IQR : ",IQR)
```

```python
#handling the outliner
stock_df.loc[stock_df['Volume']>high_vol,'Volume']=high_vol
```

```python
#starting with ML model
x=stock_df[['Open','High','Low','Adj Close','Volume']]
x
```

```python
y=stock_df[['Close']]
y
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
scale.fit_transform(x_train)
scale.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
m=reg.coef_
c=reg.intercept_
m,c
```

```python
y_predicted_train=reg.predict(x_train)
y_predicted_test=reg.predict(x_test)
y_predicted_train,y_predicted_test
```

```python
#accuracy percent 
from sklearn.metrics import r2_score
r2_s=r2_score(y_train,y_predicted_train)
r2_s
```

```python
r2_s2=r2_score(y_test,y_predicted_test)
r2_s2
```
